```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.style.use('ggplot')
df = pd.read_csv("740509512_72018_80_airline_delay_causes.csv")
df.shape
```

```python
df.loc[0,:]
```

```python
df['on_time'] = df['arr_flights'] - df['carrier_ct'] - df[' weather_ct'] - df['nas_ct'] - df['security_ct'] - df['late_aircraft_ct'] - df['arr_cancelled'] - df['arr_diverted']
df['on_time_%'] = df['on_time']/df['arr_flights']
df = df[df['year'] != 2003]
df = df[df['year'] != 2018]
df = df[np.isfinite(df['on_time_%'])]
df.shape
```

```python
grouped_data = df.groupby('year').sum()
grouped_data.head(3)
```

```python
grouped_data = df.groupby(['year', 'carrier_name']).mean()
grouped_data.head(3)
```

```python
final_df = grouped_data['on_time_%'].unstack().dropna(axis='columns')
final_df['All US Carriers'] = df.groupby(['year']).mean()['on_time_%']
```

```python
list(final_df)
```

```python
final_df = final_df.drop(['Alaska Airlines Inc.', 'ExpressJet Airlines Inc.', 'Hawaiian Airlines Inc.', 'JetBlue Airways','JetBlue Airways','SkyWest Airlines Inc.','Southwest Airlines Co.'], 1)
```

```python
#plot data
fig, ax = plt.subplots(figsize=(15,7))
final_df.plot(ax=ax, figsize=(14,8), grid=True, style='o-')
ax.legend(loc='upper center', bbox_to_anchor=(1.15, 1), shadow=True, ncol=1)
plt.ylabel('Percentage')
plt.xlabel('Year')
plt.title('On Time Performance of US Air Carrier')
```
